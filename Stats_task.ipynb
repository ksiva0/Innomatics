```python
import numpy as np
import matplotlib.pyplot as plt

class distribution:
    def __init__(self, p):
        self.p = p  # Probability of success (for relevant distributions)

    def bernoulli(self, x):
        return self.p if x == 1 else 1 - self.p

    def geometric(self, k):
        return (1 - self.p) ** (k - 1) * self.p

    def binomial(self, n, k):
        return np.math.comb(n, k) * (self.p ** k) * ((1 - self.p) ** (n - k))

    def neg_binomial(self, r, k):
        return np.math.comb(k - 1, r - 1) * (self.p ** r) * ((1 - self.p) ** (k - r))

    def hyper_geometric(self, N, K, n, k):
        return (np.math.comb(K, k) * np.math.comb(N - K, n - k)) / np.math.comb(N, n)

    def poisson(self, lam, k):
        return (np.exp(-lam) * lam ** k) / np.math.factorial(k)

    def plot_pmf(self, pmf_vals, title, position):
        plt.subplot(2, 3, position)
        plt.stem(range(len(pmf_vals)), pmf_vals)
        plt.xlabel("X")
        plt.ylabel("PMF")
        plt.title(title)


if __name__ == "__main__":
    p_success = 0.3
    custom_dist = distribution(p_success)

    
```

```python
plt.figure(figsize=(12, 8))

# Bernoulli PMF
x_vals = [0, 1]
bernouli_vals = [custom_dist.bernoulli(x) for x in x_vals]
custom_dist.plot_pmf(bernouli_vals, "Bernoulli", 1)

# Geometric PMF
k_vals = range(1, 11)
geometric_vals = [custom_dist.geometric(k) for k in k_vals]
custom_dist.plot_pmf(geometric_vals, "Geometric", 2)

# Binomial PMF
n_trials = 10
binomial_vals = [custom_dist.binomial(n_trials, k) for k in range(n_trials + 1)]
custom_dist.plot_pmf(binomial_vals, "Binomial", 3)

# Negative Binomial PMF
r_successes = 3
neg_binomial_vals = [custom_dist.neg_binomial(r_successes, k) for k in range(1, 11)]
custom_dist.plot_pmf(neg_binomial_vals, "Negative Binomial", 4)

# hyper_geometric PMF
N_total = 50
K_successes = 10
n_draws = 5
hyper_geometric_vals = [custom_dist.hyper_geometric(N_total, K_successes, n_draws, k) for k in range(n_draws + 1)]
custom_dist.plot_pmf(hyper_geometric_vals, "hyper_geometric", 6)
    
# Poisson PMF
lam = 3
poisson_vals = [custom_dist.poisson(lam, k) for k in range(11)]
custom_dist.plot_pmf(poisson_vals, "Poisson", 5)

plt.tight_layout()
plt.show()
```
